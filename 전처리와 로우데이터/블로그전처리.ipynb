```python
from konlpy.tag import Komoran
from konlpy.tag import Okt
import glob
import pandas as pd
import re
import nltk
from konlpy.corpus import kobill
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
from konlpy.tag import Okt
```

```python
keyword = '팬스타크루즈'
```

```python
file_list=glob.glob('./data/insta/*.csv')
```

```python
file_list
```

```python
data = pd.read_csv('./data/insta/naverblog_팬스타크루즈.csv',engine='python',encoding='utf-8',names=['text'])
```

```python
advertise = ['구두','문의','전화','후원','전문','도배']
a = pd.unique(data['text'])
text_list=[]
for text in a:
    flag=False
    for ad_text in advertise:
        if text.find(ad_text) >= 0:
            flag=True
            break
    if flag == False:               
        text_list.append(text)  
```

```python
text_list
```

```python
p_t = Okt(max_heap_size=12096)
p_tokens_ko = p_t.nouns(text_list)
p_ko = nltk.Text(p_tokens_ko, name = keyword)
```

```python
###오리지널
#stop_words = [keyword,'팬','스타','크루즈']
###
stop_words = [keyword,'팬','출발','월','호','수','','일','항','년','스타','크루즈','부산','오사카','여행','일본','바다','배','또','타고','그램','오늘','한번','때','박','배타','번','오후','층']
#['검색','가입','강남','다음','하나','진짜','창업','코엑스']
p_tmp        = [ word for word in p_ko if word not in stop_words ]
p_ko = nltk.Text(p_tmp, name= keyword)
p_data = p_ko.vocab().most_common(200)
```

```python
freq = dict(p_data)
```
