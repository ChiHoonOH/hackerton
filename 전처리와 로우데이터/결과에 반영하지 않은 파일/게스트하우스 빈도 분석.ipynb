```python
advertise = ['λ§μΌ€ν…','μµμ‹ ','νƒ€κ²','λ…Έμ¶','κ²μ‹','λ„λ°°','μμ§‘','κ½ƒμ§‘','μ›λ“','μ„ νƒ','ν¨κ³Ό','λ™μ•','μ„λΉ„μ¤','μΈκΈ°','κ°€λ¥','μ—…μΆ…','μ—°λ½μ²','μ‘μ„±','ν…','κΈ°λ²•','λ€λ‰','λ¤','λ‡μ¥','λ°°ν¬','μ›κ³ ','μ§€μ†','λ¬Έμ']
a = pd.unique(data['text'])
text_list=[]
for text in a:
    flag=False
    for ad_text in advertise:
        if text.find(ad_text) >= 0:
            flag=True
            break
    if flag == False:               
        text_list.append(text)  
            
            
```

```python
def stripe(st):
    RE_EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
    return RE_EMOJI.sub(r'', st)
```

```python
p_doc = ''.join(text_list)
```

```python
p_doc = stripe(p_doc)
```

```python
p_t = Okt(max_heap_size=15096)
p_tokens_ko = p_t.nouns(p_doc)
p_ko = nltk.Text(p_tokens_ko, name = keyword)
```

```python

#μ›λ³Έ
#stop_words = [keyword,'νΈμΊ‰μ¤']
stop_words = [keyword,'νΈμΊ‰μ¤','λ°','μΈμ¤νƒ€κ·Έλ¨','ν‚¤','μμ‹­','μ‚¬','μ','λ³΄κ³ ','ν•΄μ‹νƒκ·Έ','μ μ£Ό','κ·Έλ¨','μΈμ¤νƒ€','λ‚ ']


#['λ§μΌ€ν…','μµμ‹ ','νƒ€κ²','λ…Έμ¶','κ²μ‹','λ„λ°°','μμ§‘','κ½ƒμ§‘','μ›λ“','μ„ νƒ','ν¨κ³Ό','λ™μ•','μ„λΉ„μ¤','μΈκΈ°','κ°€λ¥','μ—…μΆ…','μ—°λ½μ²','μ‘μ„±','ν…','κΈ°λ²•','λ€λ‰','λ¤','λ‡μ¥','λ°°ν¬','μ›κ³ ','μ§€μ†','λ¬Έμ']
p_tmp        = [ word for word in p_ko if word not in stop_words ]
p_ko = nltk.Text(p_tmp, name= keyword)
p_data = p_ko.vocab().most_common(200)
```

```python
dict(p_data)
```

```python
import pandas as pd
freq = dict(p_data)
df = pd.DataFrame.from_dict(freq, orient='index',columns=['freq'])
df.to_excel('./data/csv/{0}2.xlsx'.format(keyword),encoding='utf-8')
```
