```python
# 광고제거
advertise = ['마케팅','최신','타겟','노출','게시','도배','수집','꽃집','워드','선택','효과','동안','서비스','인기','가능','업종','연락처','작성','팅','기법','대량','덤','몇장','배포','원고','지속','문의']
adplus = ['스타','게스트','기록','예약','제주도','하나','맞팔','선팔','문의','제공','제주시'
         ,'위치','카톡','피드','이벤트','여러분','샴푸','린스','바디','워시','안제','드라이기','수건','무한','온돌방','소규모','문자','전화','가능','주소','한림읍','협재리',
         '환영','추천','팔로우','인친','하나뿐인'
]
advertise.extend(adplus)
a = pd.unique(data['text'])
text_list=[]
for text in a:
    flag=False
    for ad_text in advertise:
        if text.find(ad_text) >= 0:
            flag=True
            break
    if flag == False:               
        text_list.append(text)    
            
            
    
            
            
```

```python
def stripe(st):
    RE_EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
    return RE_EMOJI.sub(r'', st)
```

```python
p_doc = ''.join(text_list)
```

```python
text_list
```

```python
p_doc = stripe(p_doc)
```

```python
p_t = Okt(max_heap_size=15096)
p_tokens_ko = p_t.nouns(p_doc)
p_ko = nltk.Text(p_tokens_ko, name = keyword)
```

```python

#원본
#stop_words = [keyword,'호캉스']
# 불용어 제거
stop_words = [keyword,'호캉스','및','인스타그램','키','수십','사','수','보고','해시태그','제주','그램','인스타','날']
stop_plus = ['여행','공간','협재','오라','위','옥','협재바다','곳','오늘','더','하루','번','유럽','최고','가장','움','함','비','자리','저','가지','데','것','듯','니','담','게','임','여행자','휴식','우리','주인','정서',
            '할머니','한눈','자율','진짜','가요','일리','때','또','내','시간','일기','여기','나','이','요','이번','꼭','정말'
            ,'자주','오','다시','날씨','거','셀카','겨울','그','페소','눈','한국','용','이제','스냅','도쿄','제','전',
            '이유','완전','생각','서울','너','기분','계획','어디','투어','왜','길','글','앞','처음','난','바',
            '좀','지금','유카타','첫','볼','한번','홍콩','국내','가야','중','아침','그림','샷','뭐','인','밥','세상','대박','다른',
            '맘','건물','유디','언제','카드','커피','시작','영국','시베리아','도','거리','제일','속','여행지','후쿠오카',
            '사용','줄']
stop_words.extend(stop_plus)
#['마케팅','최신','타겟','노출','게시','도배','수집','꽃집','워드','선택','효과','동안','서비스','인기','가능','업종','연락처','작성','팅','기법','대량','덤','몇장','배포','원고','지속','문의']
p_tmp        = [ word for word in p_ko if word not in stop_words ]
p_ko = nltk.Text(p_tmp, name= keyword)
p_data = p_ko.vocab().most_common(200)
```
